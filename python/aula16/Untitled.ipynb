```python
reviews_df = pd.read_csv('archive\\olist_order_reviews_dataset.csv')
```

```python
reviews_df = reviews_df[reviews_df['review_comment_message'].notna()]
```

```python
reviews_df = reviews_df[['review_score', 'review_comment_message']]
```

```python
for i, row in reviews_df.iterrows():
    score = row['review_score']
    comment = row['review_comment_message']
    print(score)
    print(comment)
    print()
    if i > 100:
        break
```

```python
reviews_df['is_positive'] = reviews_df['review_score'].isin([4, 5])
```

```python
stop_words = ['o', 'a', 'no', 'na', 'do', 'da']
def process_text(text):
    text = text.lower()
    text = text.strip()
    words = [word for word in text.split() if word not in stop_words]
    return words
```

```python
def count_words_in_texts(texts):
    counter = dict()
    for text in texts:
        words = process_text(text)
        for word in list(set(words)):
            count = counter.get(word, 0)
            counter[word] = count + 1
    counter_list = [(k, v/len(texts)) for k, v in counter.items()]
    counter_list = sorted(counter_list, key=lambda x: -x[1])
    return counter_list
```

```python
positive_texts = reviews_df[reviews_df['is_positive']]['review_comment_message'].tolist()
negative_texts = reviews_df[~reviews_df['is_positive']]['review_comment_message'].tolist()
```

```python
len(positive_texts) / (len(negative_texts) + len(positive_texts))
```

```python
positive_words = count_words_in_texts(positive_texts)
```

```python
negative_words = count_words_in_texts(negative_texts)
```

```python
vocabulary = [x[0] for x in positive_words] + [x[0] for x in negative_words]
vocabulary = list(set(vocabulary))
```

```python
positive_words_map = dict(positive_words)
negative_words_map = dict(negative_words)
```

```python
diff_words = []
for word in vocabulary:
    positive_perc = positive_words_map.get(word, 0)
    negative_perc = negative_words_map.get(word, 0)
    diff_perc = positive_perc - negative_perc
    diff_words.append((word, diff_perc))
```
